```python
#remove all rows that have invalid numbers and/or impute missing numbers
df_filtered = df[
        (df['Unnamed_13'].isna() == False) |
        (df['Unnamed_14'].isna() == False) |
        (df['Unnamed_15'].isna() == False) |
        (df['Unnamed_16'].isna() == False) |
        (df['country'] == '0')
        ]
```

```python
def cleanDataFrame(dataframe):

    for index, row in dataframe.iterrows():

        val16 = str(row.Unnamed_16)
        val15 = str(row.Unnamed_15)
        val14 = str(row.Unnamed_14)
        val13 = str(row.Unnamed_13)

        cols = list(dataframe)

        if val16 != 'nan':
            project_name = row[cols[0]] + row[cols[1]] + row[cols[2]] + row[cols[3]] + row[cols[4]]
            project_name = project_name.replace('  ',' ').strip()
            df_filtered.at[index, cols[0]] = project_name
            for i in range(1,12):
                j = i + 4
                df_filtered.at[index, cols[i]] = row[cols[j]]
            for i in range(12,16):
                df_filtered.at[index,cols[i]] = np.nan

        if val15 != 'nan' and val16 == 'nan':
            project_name = row[cols[0]] + row[cols[1]] + row[cols[2]] + row[cols[3]]
            project_name = project_name.replace('  ',' ').strip()
            df_filtered.at[index, cols[0]] = project_name
            for i in range (1,12):
                j = i + 3
                df_filtered.at[index, cols[i]] = row[cols[j]]
            for i in range(12,15):
                df_filtered.at[index,cols[i]] = np.nan

        if val14 != 'nan' and val15 == 'nan':
            project_name = row[cols[0]] + row[cols[1]] + row[cols[2]]
            project_name = project_name.replace('  ',' ').strip()
            df_filtered.at[index, cols[0]] = project_name
            for i in range (1,12):
                j = i + 2
                df_filtered.at[index, cols[i]] = row[cols[j]]
            for i in range(12,14):
                df_filtered.at[index,cols[i]] = np.nan

        if val13 != 'nan' and val14 == 'nan':
            project_name = row[cols[0]] + str(row[cols[1]])
            project_name = project_name.replace('  ',' ').strip()
            df_filtered.at[index, cols[0]] = project_name
            for i in range (1,12):
                j = i + 1
                df_filtered.at[index, cols[i]] = row[cols[j]]
            for i in range(12,13):
                df_filtered.at[index,cols[i]] = np.nan
                
        if val13 == 'nan' and val14 == 'nan':
            project_name = row[cols[0]] + str(row[cols[1]])
            project_name = project_name.replace('  ',' ').strip()
            df_filtered.at[index, cols[0]] = project_name
            for i in range (1,12):
                j = i + 1
                df_filtered.at[index, cols[i]] = row[cols[j]]
                
```
